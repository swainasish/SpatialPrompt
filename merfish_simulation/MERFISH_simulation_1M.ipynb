```python
import scanpy as sc
import pandas as pd
import numpy as np
import spatialprompt as sp
```

# import the data

```python
merfish_h5ad = sc.read_h5ad("/media/swainasish/main_disk_2/Datasets_swain_asish/spatial_prompt/reproduce_results/merfish_full.h5ad")
merfish_h5ad
```

## simulate 100 k data and run spatialprompt

```python
indexes=np.random.choice(119927,100000)
merfish_h5ad_100k=merfish_h5ad[indexes,:]
merfish_h5ad_100k
```

```python
clus_model = sp.SpatialCluster()
merfish_arr=merfish_h5ad_100k.X.toarray()
merfish_clus=clus_model.fit_predict(merfish_arr, merfish_h5ad_100k.obs.x, merfish_h5ad_100k.obs.y,
                                    n_cluster=25,n_neighbor=60)
merfish_h5ad_100k.obs.loc[:,"SpatialPrompt"]=merfish_clus
sc.pl.spatial(merfish_h5ad_100k,color="SpatialPrompt",spot_size=0.04,palette="tab20")
```
